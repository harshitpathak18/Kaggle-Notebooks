## <div style="background:#8BBB06; color: white; font-size: 25px; text-align: left; width: auto; height: auto; padding:10px; padding-left:20px; margin:2px;text-shadow: 2px 2px 5px rgba(1, 1, 1, 0.4); border-radius:50px; @import url('https://fonts.googleapis.com/css2?family=Akaya+Telivigala&display=swap'); font-family: 'Akaya Telivigala', cursive;">Data Preprocessing</div>


```python
# Initialize the LabelEncoder
encoder = LabelEncoder()

# Apply LabelEncoder to each categorical column in the DataFrame
df_encoded = df.copy()  # Make a copy of the original DataFrame to preserve it
for column in df_encoded.select_dtypes(include=['object', 'category']).columns:
    df_encoded[column] = encoder.fit_transform(df_encoded[column])

# Now df_encoded contains the encoded values
df_encoded.head()
```

```python
# feature scaling
df_scaled = df_encoded.copy()
scaler = StandardScaler()
scaling_feature = ["Age", "Salary", "Wife Salary", "Total Salary", "Price"]
df_scaled[scaling_feature]  = scaler.fit_transform(df_scaled[scaling_feature])
```

## <div style="background:#8BBB06; color: white; font-size: 25px; text-align: left; width: auto; height: auto; padding:10px; padding-left:20px; margin:2px;text-shadow: 2px 2px 5px rgba(1, 1, 1, 0.4); border-radius:50px; @import url('https://fonts.googleapis.com/css2?family=Akaya+Telivigala&display=swap'); font-family: 'Akaya Telivigala', cursive;">Market Segmentation</div>


```python
wcss = []
ss = []

# Perform PCA
pca = PCA(2)
pca_encoded = pca.fit_transform(df_scaled)

for i in range(2,12):
    kmean = KMeans(n_clusters=i, n_init='auto', random_state=42)
    kmean.fit(df_scaled)
    wcss.append(kmean.inertia_)
    
    y = kmean.predict(df_scaled)
    s = silhouette_score(pca_encoded, y, random_state=0)
    ss.append(round(s, 5))
```

```python
# Identifying Optimal Value of K for KMeans Clustering
fig, axs = plt.subplots(1,2, figsize=(15, 4))
sns.pointplot(x=list(range(2, 12)), y=wcss, color='Green', ax = axs[0])
axs[0].set_title("Elbow Plot", size=15)
axs[0].set_xlabel("Number of Clusters", size=12)

sns.pointplot(x=list(range(2, 12)), y=ss, ax=axs[1], color='Teal')
axs[1].set_title("Silhouette Plot", size=15)
axs[1].set_xlabel("Number of Clusters", size=12)
plt.show()
```

<div style="background:#84D19B; color: white; font-size: 16px; text-align: left; width: auto; height: auto; padding:10px; padding-left:20px; margin:2px;text-shadow: 2px 2px 5px rgba(1, 1, 1, 0.4); border-radius:10px; @import url('https://fonts.googleapis.com/css2?family=Akaya+Telivigala&display=swap'); font-family: 'Akaya Telivigala', cursive;">
Note - Both the elbow and silhouette methods indicate that three clusters are optimal for the K-means algorithm.</div>

```python
# Apply KMeans with 4 clusters and plotting Clusters Using PCA
Clustering = KMeans(n_clusters=4, random_state=2).fit(df_scaled)
labels = Clustering.labels_
df['Cluster'] = labels
df_encoded['Cluster'] = labels

# Perform PCA for 2D visualization
pca = PCA(n_components=2)
pca_df = pca.fit_transform(df_scaled)

# Transform the cluster centers using the same PCA transformation
centers = pca.transform(Clustering.cluster_centers_)

# 2D Scatter Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot( x=pca_df[:, 0], y=pca_df[:, 1], hue=labels, palette="Set1", legend='full',alpha=0.9)
plt.scatter(centers[:, 0], centers[:, 1], c='#A613F5', s=100, marker='*', label='Centroids')
plt.title("KMeans Clustering with 4 Clusters (2D)", size=15)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.grid(True)
plt.show()

# Perform PCA for 3 components for 3D plotting
pca_3d = PCA(n_components=3)
pca_df_3d = pca_3d.fit_transform(df_scaled)
centers_3d = pca_3d.transform(Clustering.cluster_centers_)

fig = px.scatter_3d(x=pca_df_3d[:, 0],y=pca_df_3d[:, 1],z=pca_df_3d[:, 2],color=labels,title="KMeans Clustering with 4 Clusters (3D)",
    labels={'x': 'PCA Component 1', 'y': 'PCA Component 2', 'z': 'PCA Component 3'})

# Add cluster centroids to the 3D plot
fig.add_scatter3d(
    x=centers_3d[:, 0],
    y=centers_3d[:, 1],
    z=centers_3d[:, 2],
    mode='markers',
    marker=dict(size=5, color='red', symbol='x'),
    name='Centroids'
)

fig.show()
```

```python
data_pivot = df_encoded.groupby('Cluster').mean().T
data_pivot_mean = data_pivot.mean(axis=1).reset_index()
data_pivot_mean.columns = ['Variable', 'Value']
```

```python
# Segment profile plot for the four-segment
plt.figure(figsize = (12, 9))
for i in range(4):
    plt.subplot(2, 2, i+1)
    sns.barplot(data_pivot, x = i, y = data_pivot.index, palette='viridis')
    sns.scatterplot(data_pivot_mean, x = 'Value', y = 'Variable', color = 'green')
    for index, row in data_pivot_mean.iterrows():
        plt.hlines(y=row['Variable'], xmin=0, xmax=row['Value'], colors='black', alpha = 0.8)
    plt.ylabel("")
    plt.xlabel("")
    plt.title(f"Cluster {i}: {df_encoded['Cluster'].value_counts()[i]} ({df_encoded['Cluster'].value_counts()[i]*100/len(df_encoded):.0f}%)")

plt.suptitle("Segment profile plot for the four-segment solution", fontsize = 15)
plt.tight_layout()
plt.show()
```

<h4 style="color:red">Cluster 0 (High-Income Families):
This segment is characterized by older, high-income families with multiple dependents. They typically have a substantial combined household income and prefer to invest in higher-priced vehicles, such as SUVs. The majority are married, well-educated, and have stable professional careers. Notably, both partners are likely to be working, and they are less reliant on personal or house loans.
</h4>

<h4 style="color:skyblue">Cluster 1 (Emerging Young Professionals):
This group consists mainly of younger, single professionals with fewer dependents. Their income levels are moderate, and their car purchases reflect this, with a preference for mid-range vehicles like the Baleno. This cluster is in the early stages of their careers, with many still unmarried, and they tend to avoid taking out loans. Their educational background is strong, but they have not yet reached the peak of their earning potential.
</span>

<h4 style="color:green">Cluster 2 (Affluent Young Professionals):
These individuals are slightly older than those in Cluster 1 and are typically married with a couple of dependents. They have a solid income and often choose cars in the mid-range price bracket. Despite their affluence, they do not commonly rely on loans and are in stable, salaried positions. Their level of education is high, and they generally prefer not to have both partners working.
</span>
   
<h4 style="color:purple">Cluster 3 (Family-Oriented Middle-Income):
Cluster 3 consists of middle-income families with a higher number of dependents. They have a moderate household income and prefer vehicles that balance cost and quality, like the Baleno. They are mostly married, with a good educational background, and are likely to have taken out house loans. The working status of the wife is typically non-working, reflecting a traditional family structure.
</h4>
